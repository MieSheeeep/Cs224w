# 导入工具包

```python
import networkx as nx
# 数据可视化
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei'] # 设置中文字体为黑体
plt.rcParams['axes.unicode_minus'] = False # 解决负号'-'显示为方块的问题
```

# 经典图结构

```python
# 全连接无向图
G=nx.complete_graph(4)
nx.draw(G)
# 全图连接数
print(G.size())
```

```python
# 全连接有向图
G=nx.complete_graph(4, create_using=nx.DiGraph())
nx.draw(G)
print(G.is_directed())
```

```python
# 环状图
G=nx.cycle_graph(6)
nx.draw(G)
```

```python
# 梯度图
G=nx.ladder_graph(5)
nx.draw(G)
```

```python
# 线性串珠图
G=nx.path_graph(6)
nx.draw(G)
```

```python
# 星状图
G=nx.star_graph(5)
nx.draw(G)
```

```python
# 轮辐图
G=nx.wheel_graph(6)
nx.draw(G)
```

```python
# 二项树
G=nx.binomial_tree(5)
nx.draw(G)
```

```python
# 栅格图
G1=nx.grid_2d_graph(3,5)
nx.draw(G1)
```

```python
G2=nx.grid_graph(dim=[2,3,4])
nx.draw(G2)
```

```python
G3=nx.triangular_lattice_graph(2,5)
nx.draw(G3)
```

```python
G4=nx.hexagonal_lattice_graph(2,3)
nx.draw(G4)
```

```python
G5=nx.hypercube_graph(4)
nx.draw(G5)
```

```python
# Petersen图
G=nx.petersen_graph()
nx.draw(G)
```

```python
# 风筝图
G=nx.krackhardt_kite_graph()
nx.draw(G)
```

## 有向图

```python
# 无标度有向图
G=nx.scale_free_graph(10) # 有概率产生多重边和自环
nx.draw(G)
```

## 其他常用图

```python
# 社交网络
G=nx.karate_club_graph() # 空手道俱乐部社交网络
nx.draw(G,with_labels=True)
```

```python
# 查看节点类别
print(G.nodes[20]['club'])
print(G.nodes[21]['club'])
```

```python
# 雨果悲惨世界人物关系图
G=nx.les_miserables_graph()
plt.figure(figsize=(12,10))
pos=nx.spring_layout(G,seed=10) # 设置布局
nx.draw(G,pos,with_labels=True)
```

```python
# 家庭谱系图
G=nx.florentine_families_graph()
nx.draw(G,with_labels=True)
```

```python
# 社群聚类图
G=nx.caveman_graph(4,3)
nx.draw(G,with_labels=True)
```
